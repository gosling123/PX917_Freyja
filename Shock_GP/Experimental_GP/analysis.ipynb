```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import pickle
import csv
import glob
plt.rcParams["figure.figsize"] = (10,5)
```

## Check Experimental File


```python
# Import experimental datasets

X = {}; 
Y = {}; 
err = {}; 

with open('experimental_GP_data.pickle', 'rb') as f:
    X[1],Y[1],err[1],X[2],Y[2],err[2],X[3],Y[3],err[3],X[4],Y[4],err[4]= pickle.load(f)

```

```python
labels = {}

labels[1] = 'Fig 2' ; labels[2] = 'Fig 3a'; labels[3] = 'Fig 3b'; labels[4] = 'Fig 4'

for i in range(1, 5):
    plt.plot(X[i], Y[i], label = labels[i])
    plt.fill_between(X[i], Y[i]-err[i], Y[i]+err[i], alpha = 0.5)
    # plt.xlim(3,4)
    if i == 1:
        plt.ylabel(r' Breakout Time (ns)')
        plt.xlabel(r'Thickness ($\mu$ m)')
    else:
        plt.ylabel(r'$U_{shock} (\mu m/ns) $')
        plt.xlabel(r'Time (ns)')

    plt.show()
```
